```python
# Import the neccesary libraries needed for this project
import opendatasets as od
import os
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
print('Libraries succesfully imported')
```

```python
# Download the police shooting ".csv" dataset from the kaggle website
dataset_url = 'https://www.kaggle.com/ahsen1330/us-police-shootings'
od.download(dataset_url)

data_dir = './us-police-shootings'
os.listdir(data_dir)
```

```python
# Read downloaded data into dataframe and diplaay the first five rows of the dataset
shooting_df = pd.read_csv(data_dir +'/shootings.csv')
shooting_df.head()
```

```python
# Check the attributes of all columns in the heart_df dataframe using the info method

shooting_df.info()
```

```python
# Check for missing data
shooting_df.isna().sum()
```

<h3 style = 'color: #0A00A6;'> Data Type Conversion</h3>

There are no null values in all columns of this dataset. However, it is only right to convert columns to a more suitable data type for the sake of a better and faster analysis

```python
# Convert the date column to datetime data type
shooting_df['date'] = pd.to_datetime(shooting_df['date'])

# Convert the age column to integers
shooting_df['age'] = shooting_df['age'].astype('int')

# Convert the following columns to category data type as it helps reduce memory usage
shooting_df['manner_of_death'] = shooting_df['manner_of_death'].astype('category')
shooting_df['armed'] = shooting_df['armed'].astype('category')
shooting_df['threat_level'] = shooting_df['threat_level'].astype('category')
shooting_df['gender'] = shooting_df['gender'].astype('category')
shooting_df['race'] = shooting_df['race'].astype('category')
shooting_df['arms_category'] = shooting_df['arms_category'].astype('category')
shooting_df['flee'] = shooting_df['flee'].astype('category')

# It is only proper for the gender column to display genders in full i.e F = Female and M = Male hence the need to replace accordingly 
shooting_df['gender'] = shooting_df['gender'].replace(to_replace = ['F', 'M'], value = ['Female', 'Male'])
```

```python
# Dictionary of abbreviation of US states to full name
us_state_abbrev = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AS': 'American Samoa', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'DC': 'District of Columbia', 'FL': 'Florida',
    'GA': 'Georgia', 'GU': 'Guam', 'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa',
    'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland', 'MA': 'Massachusetts',
    'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi', 'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska',
    'NV': 'Nevada', 'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina',
    'ND': 'North Dakota', 'MP': 'Northern Mariana Islands', 'OH': 'Ohio', 'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania',
    'PR': 'Puerto Rico', 'RI': 'Rhode Island', 'SC': 'South Carolina', 'SD':'South Dakota', 'TN':'Tennessee', 'TX':'Texas',
    'UT': 'Utah', 'VT': 'Vermont', 'VI': 'Virgin Islands', 'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia',
    'WI': 'Wisconsin', 'WY': 'Wyoming'}

# Create a new colunmn to show the full name of US state
shooting_df['state_full'] = shooting_df['state'].map(us_state_abbrev)

# let us frop the ID column as they won't be useful in this project
shooting_df.drop(['id', 'state'], axis = 1, inplace = True)
```

```python
# Using the date column, threcreate three more columns namely 'year', 'month' and 'day'
shooting_df['year'] = shooting_df['date'].dt.year
shooting_df['month'] = shooting_df['date'].dt.month
shooting_df['day'] = shooting_df['date'].dt.day
```

```python
# Set style for visualisation
sns.set_style('darkgrid')

matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (15, 10)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
print('Visualization theme set')
```

<h3 style = 'color: #0A00A6;'> 1. How many men and women have been victims of police shootings</h3>


```python
# Total count of victims by their gender
fig, ax = plt.subplots()

gender = shooting_df['name'].groupby(shooting_df['gender']).count()
sex = gender.index
counts =  gender.values

# Let red represent female and blue represent male
bar_labels = ['red', 'blue']
bar_colors = ['tab:red', 'tab:blue']

# The bar plot
hbars = ax.bar(sex, counts, label=bar_labels, color=bar_colors)
# Insert label to the plot
ax.bar_label(hbars)

ax.set_ylabel('Number of Victims')
ax.set_xlabel('Gender')
ax.set_title('Number of police shooting victim by gender')

print(f'The total count of female and male shooting victims are {gender.values[0]} and {gender.values[1]} respectively')
print(gender)
```

<h3 style = 'color: #0A00A6;'> Answer</h3><br>
The total count of female and male shooting victims are 222 and 4673 respectively

<h3 style = 'color: #0A00A6;'>2. What race has the highest number of police shooting victims?</h3>

```python
race = shooting_df['race'].groupby(shooting_df['race']).count().sort_values(ascending = True)

fig, ax = plt.subplots()

race_x = race.index
race_y =  race.values

# The bar plot
race_bars = ax.barh(race_x, race_y)
# Insert label to the plot
ax.bar_label(race_bars, padding = 5)

ax.set_ylabel("Victim's Race")
ax.set_xlabel('Number of Victims')
ax.set_title('Number of police shooting victim by race')

print(f'The race with the highest number of shooting victims is {race.index[-1]} race,  with a total of {race.values[-1]} victims')
print(race.sort_values(ascending = False))
```

<h3 style = 'color: #0A00A6;'> Answer</h3><br>
The race with the highest number of victims is the <b>white race</b> with a total of <b>2476</b> victims

<h3 style = 'color: #0A00A6;'>3. Which city has the highest case of police shooting?</h3>

```python
state = shooting_df['city'].groupby(shooting_df['state_full']).count().sort_values(ascending = False).head(20)

plt.figure(figsize = (15,15))

state_plot = sns.barplot(x= state.values, y= state.index)
state_plot.bar_label(state_plot.containers[0], padding = 5)


plt.title('State distribution of police shooting victims')
plt.xlabel('Victim Count')
plt.ylabel('State of victims');

print(state)
```

<h3 style = 'color: #0A00A6;'> Answer</h3><br>
The state with the highest number of victims is <b>Carlifonia</b> with a total of <b>701</b> victims between 2015 to 2020

<h3 style = 'color: #0A00A6;'>4. How many shooting victims were diagnozed with mental illness?</h3>

```python
mental = shooting_df['signs_of_mental_illness'].groupby(shooting_df['signs_of_mental_illness']).count()

mental_y = mental.index

ment = mental.plot(kind = 'pie', y = mental_y, autopct='%1.0f%%',  colors = ['pink', 'orange'],
                    title='Number of victims diagnosed with mental illness');

print(mental)
```

```python
# Create the data required for grouping

grouped = shooting_df.groupby([shooting_df['gender'],shooting_df['signs_of_mental_illness']]).size().unstack(fill_value=0)
grouped_plot = grouped.plot.bar()
 
for container in grouped_plot.containers:
    grouped_plot.bar_label(container)

plt.title('Mental illness of Victims by Gender')
plt.xlabel('Mental illness sign by sex')
plt.ylabel('Victim count')
plt.legend(title = 'Sign of mental illness');

grouped
```

<h3 style = 'color: #0A00A6;'>7. How many victims attacked with a weapon and which weapon was mostly used by victims?</h3>

```python
# Create a subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,15), gridspec_kw={'width_ratios': [5, 5]})

#Filter for weapon data
attack = shooting_df['armed'].groupby(shooting_df['armed']).count().nlargest(10).sort_values(ascending = True)
weapon = shooting_df['arms_category'].groupby(shooting_df['arms_category']).count().sort_values(ascending = True)

# Give the subplot a title
fig.suptitle('Count of victims armed with weapons vs most used weapon')


#Plot the first subplot
weapon_plot = ax1.barh(attack.index, attack.values)
ax1.bar_label(weapon_plot, padding = 5)
ax1.set_title('Ten most used weapons used by victims')
ax1.set_xlabel('Victim count')
ax1.set_ylabel('Weapon used')

# Plot second plot
weapon2 = ax2.barh(weapon.index, weapon.values)
ax2.bar_label(weapon2, padding = 5)
ax2.set_title('Arm category')
ax2.set_xlabel('Victim count')
ax2.set_ylabel('Arm category');
```

The most used weapon is the gun under the guns category
